Question 1

```python
def chara(s):
    result = ""
    seen_chars = set()
    for char in s:
        if char not in seen_chars:
            count = s.count(char)
            result += f"{char}={count}, "
            seen_chars.add(char)
    result = result.rstrip(", ")
    return result
instr= input("Enter a string value: ")
freq=chara(instr)
print (f"Occurrence of all characters in: {instr} \n "+ str(freq))
```

```python
def exponent(base, expo):
    result = 1
    for i in range(expo):
        result *= base
    return result
base = int(input("Enter the base: "))
expo = int(input("Enter the exponent: "))
print("Result:", exponent(base, expo))
```

Question 4

```python
def cube(n):
    sum = 0
    for i in range(1, n):
        sum+= i ** 3
    return sum
n = int(input("Enter a positive integer: "))
result = cube(n)
print(f"Sum of cubes of positive integers smaller than {n} is: {result}")
```

```python
def squares(numbers):
    sum = 0
    for num in numbers:
        sum += num ** 2
    return sum
d=int(input('What is the size of list '))
t=[]
su=0
for i in range(d):
  num=int(input('Number= '))
  t.append(num)
print(t)
result = squares(t)
print("Sum of squares of the numbers in the list is:", result)
```

Question 9

```python
def fahrenheit(celsius):
    fahrenheit = (celsius * 9/5) + 32
    return fahrenheit
def celsius(fahrenheit):
    celsius = (fahrenheit - 32) * 5/9
    return celsius
choice = input("Enter 'F' for Fahrenheit, or 'C' for Celsius: ")
if choice == 'F':
    celsius = float(input("Enter temperature in Celsius: "))
    fahrenheit = fahrenheit(celsius)
    print(f"{celsius} Celsius is equal to {fahrenheit} Fahrenheit")
elif choice == 'C':
    fahrenheit = float(input("Enter temperature in Fahrenheit: "))
    celsius = celsius(fahrenheit)
    print(f"{fahrenheit} Fahrenheit is equal to {celsius} Celsius")
else:
    print("Invalid choice")
```

```python
def fact(n):
  if n < 0:
    return None
  elif n==1:
    return n
  else:
    return n*fact(n-1)
inp=int(input("Enter a number"))
print(f"Factorial of {inp} is {fact(inp)}")
```
