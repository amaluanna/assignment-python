```python
def circle_area (radius):
    pi = 3.14159
    area = pi * radius ** 2
    return area
n1=int(input('Enter the radius '))
a = circle_area(n1)
print("Area of the circle is:", a)
```

```python
def largest(num):
     largest = num[0]
     for n in num:
        if n > largest:
            largest = n
     return largest
s=int(input('What is the size of list '))
t=[]
su=0
for i in range(s):
  num=int(input('Number= '))
  t.append(num)
print(t)
l=largest(t)
print(f'{l} is the largest number')
```

```python
def sum(s,l):
    num1 = s -1
    for num2 in range(s, l +1):
        sum = num2 + num1
        print(f"Current Number {num2} Previous Number {num1} Sum: {sum}")
        num1 = num2
first=int(input('First number= '))
end=int(input('Last number= '))
sum(first,end)
```

```python
def reverse(list1):
    rlist = []
    l=len(list1)
    for i in range(l-1,-1, -1):
        rlist.append(list1[i])
    return rlist
s=int(input('What is the size of list '))
list1=[]
su=0
for i in range(s):
  num=int(input('Number= '))
  list1.append(num)
list2 = reverse(list1)
print("Original List:", list1)
print("Reversed List:", list2)
```

```python
def pattern(row):
    for i in range(1, row + 1):
        for s in range(1, i + 1):
            print('*', end='')
        print('\n')
    for i in range(row - 1, 0, -1):
        for s in range(1, i + 1):
            print('*', end='')
        print('\n')
pattern(5)
```
